# Pydantic Data Validation

```python
from langgraph.graph import StateGraph, START, END
from pydantic import BaseModel
```

```python
class State(BaseModel):
    name: str
```

```python
def example_mode(state: State):
    return {"name": "Hello"}
```

```python
# stateGraph
builder = StateGraph(State)

builder.add_node("example_mode", example_mode)

builder.add_edge(START, "example_mode")
builder.add_edge("example_mode", END)

graph = builder.compile()
graph
```

```python
graph.invoke({"name": 123})
```
